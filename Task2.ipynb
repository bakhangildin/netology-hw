# Задание 1

```python
def is_leap_year(year: int) -> bool:
    return year % 4 == 0

def solution1(year: int) -> None:
    if is_leap_year(year):
        print("Високосный год")
    else:
        print("Обычный год")
```

```python
year = 2028
solution1(year)

year = 2027
solution1(year)
```

```python
def run_tests1() -> None:
    for year in range(2000, 2026):
        print(year, end=" ")
        solution1(year)
run_tests1()
```

# Задание 2

```python
def is_ticket_lucky(number: int) -> bool:
    s = str(number)
    assert len(s) == 6, "Ticket number must contain 6 digits"
    return s[:3] == s[3:6][::-1]
    
    
def solution2(number: int) -> None:
    if is_ticket_lucky(number):
        print("Счастливый билет")
    else:
        print("Несчастливый билет")
```

```python
number = 123456
solution2(number)

number = 123321
solution2(number)
```

```python
def run_tests2() -> None:
    cases = (666, 666666, 765123, 69420, 789987, 694201337)
    for number in cases:
        try:
            print(number, end=" ")
            solution2(number)
        except Exception as e:
            print(e)

run_tests2()
```
