```python
documents = [
    {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
    {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
    {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]
directories = {
    "1": ["2207 876234", "11-2"],
    "2": ["10006"],
    "3": []
}
```

```python
import traceback

class QuitException(Exception):
    pass

def quit_handler(cmd: str):
    raise QuitException

def help_handler(cmd: str):
    if cmd != "h":
        print("Команда \"%s\" не найдена." % cmd)
    print("Доступные команды:")
    print("p: Вывод владельца документа по номеру")
    print("s: Поиск документа на полке по номеру")
    print("q: Выход")
    
def print_handler(cmd: str):
    number = input("Введите номер документа: ")
    owner = None
    for document in documents:
        if document["number"] == number:
            owner = document["name"]
    if owner is not None:
        print("Владелец документа: %s" % owner)
    else:
        print("Владелец документа: владелец не найден")
        
def search_handler(cmd: str):
    number = input("Введите номер документа: ")
    d = None
    for directoriy, documents in directories.items():
        for document_number in documents:
            if document_number == number:
                d = directoriy
                break
    if d is not None:
        print("Документ хранится на полке: %s" % d)
    else:
        print("Документ не найден.")

def main():
    handlers = {
        "q": quit_handler,
        "p": print_handler,
        "s": search_handler,
    }
    while True:
        cmd = input("Введите команду: ")
        handler = handlers.get(cmd, help_handler)
        try:
            handler(cmd)
        except QuitException:
            print("Завершение")
            break
        except Exception as e:
            print("miserable failure :(")
            print(traceback.format_exc(e))
            
main()
```
