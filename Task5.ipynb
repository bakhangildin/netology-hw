```python
import csv
import json
import time

data = {}
with open("task5-data/purchase_log.txt", encoding="utf-8") as plf:
    for i, line in enumerate(plf):
        if i == 0:
            continue
        line = line.strip()
        d = json.loads(line)
        user_id = d["user_id"]
        category = d["category"]
        if user_id not in data:
            # список потому что например у user_id "a33eee80c0" две покупки с разными категориями
            # Досуг, Бытовая техника
            data[user_id] = [category]
        else:
            data[user_id].append(category)
            
with open("task5-data/visit_log.csv", encoding="utf-8") as vlf:
    with open("task5-data/funnel.csv", "w", encoding="utf-8") as f:
        # т.к я разделяю дублирующиеся категории запятой нужно включить quoting
        w = csv.writer(f, quoting=csv.QUOTE_ALL)
        w.writerow(["user_id", "source", "categories"])
        for i, line in enumerate(vlf):
            if i == 0:
                continue
            line = line.strip()
            user_id, source = line.split(",")
            if user_id in data:
                categories = ", ".join(data[user_id])
            else:
                categories = ""
            w.writerow([user_id, source, categories])

with open("task5-data/funnel.csv", encoding="utf-8") as f:
    for i, line in enumerate(f):
        print(line.strip())
        if i > 50:
            break
```
