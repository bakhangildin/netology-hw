# Задание 1

```python
def print_middle_part(s: str):
    size = len(s)
    center = size // 2
    if size % 2 == 0:
        print(s[center-1:center+1])
    else:
        print(s[center])
```

```python
word = "test"
print_middle_part(word)

word = "testing"
print_middle_part(word)
```

```python
def run_tests1() -> None:
    cases = (
        "foo",
        "barbaz",
        "foobarbaz",
    )
    for word in cases:
        print(word, end=": ")
        print_middle_part(word)

run_tests1()
```

# Задание 2

```python
def run2(boys: "list[str]", girls: "list[str]") -> None:
    if len(boys) != len(girls):
        print("Внимание, кто-то может остаться без пары!")
        return
    boys = sorted(boys)
    girls = sorted(girls)
    print("Идеальные пары:")
    for i in range(len(boys)):
        print(boys[i], "и", girls[i])
```

```python
boys = ["Peter", "Alex", "John", "Arthur", "Richard"]
girls = ["Kate", "Liza", "Kira", "Emma", "Trisha"]
run2(boys, girls)
```
